```python
crime2012=df[df.index.str.find('2012')!=-1][:]
```

```python
crime2012
```

```python
x=crime2012['Primary Type'].value_counts().sort_values(ascending=False)[:5]
```

```python
def func(i):
    l=[]
    for j in ['01','02','03','04','05','06','07','08','09','10','11','12']:
        l.append(crime2012[(crime2012['Primary Type']==x.index[i]) & (crime2012.index.str[0:2]==j)]['Primary Type'].count())
    return(l,x.index[i])
```

```python
for i in range(5):
    m=np.arange(1,13,1)
    l,t=func(i)
    plt.plot(m,l,label=t)
plt.xlabel('months')
plt.ylabel('crimes')
plt.title('TOP 5 CRIMES IN CHICAGO IN 2012')
plt.xticks(m,['jan','feb','mar','apr','may','june','july','aug','sep','oct','nov','dec'])
plt.legend(bbox_to_anchor=(1,0.5))
```

* THEFTS ARE USUALLY HIGH DURING MONTHS OF JULY AND AUGUST
* CRIMINAL DAMAGES,ASSUALT AND BURGLARY ARE USUALLY LEAST IN FEBRUARY
* ALL TYPES OF CRIMES WERE HIGH IN MONTH OF JULY
